```python
import pandas as pd
Traindata = pd.read_csv("train.csv")
Traindata.head()
```

```python
Testdata = pd.read_csv("test.csv")
Testdata.sample(5)
```

```python
Traindata.isnull().sum()
```

```python
data = Traindata.append(Testdata)
data
```

```python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
data['Gender'] = lab.fit_transform(data['Gender'].astype("str"))
```

```python
features = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area']
```

```python
for feature in features:
    data[feature] = lab.fit_transform(data[feature].astype("str"))
    
```

```python
data['Gender'].value_counts()
```

```python
for column in data.columns:
    print(data[column].value_counts())
```

```python
import numpy as np
from numpy import random
columns = ['Married','Gender','Self_Employed']
for column in columns:
    data[column].fillna(np.random.randint(0,1),inplace=True)
```

```python
columns = ['LoanAmount','Dependents']
for column in columns:
    data[column].fillna(data[column].median(),inplace=True)
```

```python
columns = ['Loan_Amount_Term']
for column in columns:
    data[column].fillna(data[column].mean(),inplace=True)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```
