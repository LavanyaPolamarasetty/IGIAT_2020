```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("creditcard_csv.csv")
df.head()
```

```python
df.shape
```

```python
df["Class"].value_counts()
```

```python
val = df["Class"].value_counts()
val.plot(kind = 'bar')
plt.xlabel("Class")
plt.ylabel("Transactions")
plt.title("Transaction Distribution")
plt.ylim(0,10000)
plt.show()
```

```python
df[df['Class'] == "'1'"] = 1
df[df['Class'] == "'0'"] = 0
df["Class"].value_counts()
```

```python
fraud  = df[df['Class'] == 1]
```

```python
valid = df[df["Class"] == 0]
```

```python
print(len(fraud))
print(len(valid))
```

```python
outlier_fraction = len(fraud)/len(valid)
print("outlier_fraction is",outlier_fraction*100)
```

```python
df_samples = df.sample(frac = 0.1, random_state=0)
df_samples.shape
```

```python
X = df_samples[df_samples.columns[:-1]]
X.shape
```

```python
y = df_samples.Class
y.shape
```

```python
val = df_samples["Class"].value_counts()
val.plot(kind = 'bar')
plt.xlabel("Class")
plt.ylabel("Transactions")
plt.title("Transaction Distribution")
plt.ylim(0,10000)
plt.xticks(rotation='horizontal')
plt.show()
```

#### Isolation Forest

```python
from sklearn.ensemble import IsolationForest
iso_forest = IsolationForest(n_estimators=100)
```

```python
iso_forest.fit(X)
preds = iso_forest.predict(X)
```

```python
preds
```

```python
preds[preds == 1 ] = 0
preds[preds == -1 ] = 1
```

```python
len(preds)
```

```python
pd.Series(preds).value_counts()
```

```python
n_error = ( preds  != y ).sum()
```
