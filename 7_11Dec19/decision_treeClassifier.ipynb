```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
print(iris)
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris_data = pd.DataFrame(iris["data"])
iris_data
```

```python
iris_data.columns = iris["feature_names"]
iris_data
```

```python
iris_data["Target"] = iris["target"]
iris_data.sample(4)
```

```python
iris_data.shape
```

```python
iris_data["Target"].unique()
```

```python
iris.target_names
```

```python
iris_data.columns
```

```python
X = iris_data[["sepal length (cm)"]]
X.sample(4)
```

```python
y = iris_data["Target"]
y.sample(3)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
```

```python
tree_clf.fit(X,y)
```

```python
tree_clf.predict([[4.4],[6.2],[6.7]])
```

```python
tree_clf.predict_proba([[4.4]])
```

```python
tree_clf.predict_proba([[6.7]])
```

```python
tree_clf.score(X,y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = tree_clf.predict(X_test)
```

```python
tree_clf.score(X_test,y_test)
```

```python
print("Accuracy-score", accuracy_score(y_test,y_pred))
```
